## Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
# The LookupError class is a base class for KeyError and IndexError subclasses that occur when a key or index is not found during a lookup operation.
```

```python
try:
    dict = {'Name':"Avash","Mob_Num":9876543210}
    print(dict["Mail"])
except KeyError as e:
    print(e)
# the key 'mail' is not present in the dictionary dict, so a KeyError is raised.
```

```python
try:
    list = [1,3,5,7,9]
    print(list[10])
except IndexError as e:
    print(e)
# here list has 5 elements, so accessing an element beyond index 4 raises an IndexError
```

## Q6. List down some best practices for exception handling in python. 

```python
# Some best practices for exception handling in python are as follows:
#(a) Use specific exception for accurate error handling rather than using Exception superclass or only using bare except statement.
#(b) Use separate except blocks for handling multiple types of exceptions.
#(c) Use the logging module to log exceptions.
#(d) Use the else block for performing actions that should only occur when no exceptions are raised or for displaying success messages.
#(e) Use the finally block regardless of whether an exception occurred or not. 
#(f) Use custom exception to encounter an error that is specific to a application and can't be handled by the existing built-in exceptions.
```
